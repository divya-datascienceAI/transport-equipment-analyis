# FINDING THE PROFIT IN TRANSPORT EQUIPMENT BUSINESS.

```python
import numpy as np
import pandas as pd
import os
path='D:/Data sets/Transportation Equipment.csv'
df=pd.read_csv(path,na_values='NA',usecols=[0,1,2,3,4])
print(df.head())
print(df.tail())
df.describe(include='all')
```

```python

df.insert(5,'dollar_2021',(df.capital)*9.65)
print(df.head())
```

```python
df=df.sort_values(['dollar_2021'])
print(df.head())
```

FROME ABOVE:
    JUST SORTING THE VALUES ACCORDING TO THE DOLLARS IN 2021.

```python
#add=df2.capital + df2.labor
df.insert(6,'total spending',(df.capital+df.labor)/df.firms)
#print(df2.head())
df.insert(7,'output',df['value added']/df['firms'])
df.insert(8,'profit',df['output']-df['total spending'])
print(df.head())
```

```python
df=df.sort_values(['profit'],ascending=False)
print(df.head(6))
print(df.tail(6))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#path='D:/Data sets/Transportation Equipment.csv'
#df=pd.read_csv(path,na_values='NA',usecols=[0,1,2,3,4,5,6])
#df = pd.DataFrame(df, columns=[" US states", "total", "profit "])
df.plot(x="US states", y=["profit"], kind="bar", figsize=(15, 10))
#plt.barh(df.US states,df.profit)
for v,i in enumerate(round(df.profit,2)):
    plt.text(v, i,str(i),color = 'blue', fontweight = 'light')
plt.xlabel("US states")
plt.ylabel("profit in Million dollar")
plt.show()
#df2.hist(figsize=(20,10))
df.plot(x="US states", y=["profit"], kind="line", figsize=(15, 10))
plt.xlabel("US states")
plt.ylabel("profit in Million dollar")
plt.show()
```
